# Image Enhancement  An image enhcancement example using Python CV2. Original article:[Image Enhancement Techniques using OpenCV and Python](https://towardsdatascience.com/image-enhancement-techniques-using-opencv-and-python-9191d5c30d45)
    

**Step 1:** Import libraries

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

**Step 2** Load and display sample image

```python
%matplotlib inline
img_orig = cv2.imread('sample.JPG')
img_enh = cv2.imread('sample1.JPG')
images = np.concatenate((img_orig,img_enh),axis=1)
plt.imshow(images)
plt.show()
```

**Step 3:** Convert image to grayscale image

```python
gray_img_orig = cv2.cvtColor(img_orig,cv2.COLOR_BGR2GRAY)
gray_img_enh = cv2.cvtColor(img_enh,cv2.COLOR_BGR2GRAY)
```

**Step 4:** Plot grayscale intensity distribution

```python
hist_orig = cv2.calcHist(gray_img_orig,[0],None,[256],[0,256])
hist_enh = cv2.calcHist(gray_img_enh,[0],None,[256],[0,256])
plt.subplot(121)
plt.title("Original Image")
plt.xlabel('bins')
plt.ylabel("No. of pixels")
plt.plot(hist_orig)
plt.subplot(122)
plt.title("Enhanced Image")
plt.xlabel('bins')
plt.ylabel("No. of pixels")
plt.plot(hist_enh)
plt.show()
```

**Step 5:** Now we will use cv2.equalizeHist() function with the purpose of equalizing the contrast of a given grayscale image. cv2.equalizeHist() function normalises the brightness and also increases the contrast.

```python
gray_img_orig_eqhist=cv2.equalizeHist(gray_img_orig)
gray_img_enh_eqhist=cv2.equalizeHist(gray_img_enh)
hist_orig=cv2.calcHist(gray_img_orig_eqhist,[0],None,[256],[0,256])
hist_enh=cv2.calcHist(gray_img_enh_eqhist,[0],None,[256],[0,256])
plt.subplot(121)
plt.plot(hist_orig)
plt.subplot(122)
plt.plot(hist_enh)
plt.show()
```

**Step 6:** Display the Gray Scale Histogram equalized images

```python
%matplotlib inline
equhist_images = np.concatenate((gray_img_orig_eqhist,gray_img_enh_eqhist),axis=1)
print()
plt.imshow(equhist_images)
plt.show()
```

**Step 7** Contrast Limited Adaptive Histogram Equalization

```python
clahe = cv2.createCLAHE(clipLimit=40)
gray_img_orig_clahe = clahe.apply(gray_img_orig_eqhist)
gray_img_enh_clahe = clahe.apply(gray_img_enh_eqhist)
images = np.concatenate((gray_img_orig_clahe,
                         gray_img_enh_clahe),axis=1)
plt.imshow(images)
plt.show()
```

**Step 8:** Thresholding Techniques

```python
th=80
max_val=255
ret, o1 = cv2.threshold(gray_img_orig_clahe, th, max_val, cv2.THRESH_BINARY)
cv2.putText(o1,"Thresh_Binary",(40,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),3,cv2.LINE_AA)
ret, o2 = cv2.threshold(gray_img_orig_clahe, th, max_val, cv2.THRESH_BINARY_INV)
cv2.putText(o2,"Thresh_Binary_inv",(40,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),3,cv2.LINE_AA)
ret, o3 = cv2.threshold(gray_img_orig_clahe, th, max_val, cv2.THRESH_TOZERO)
cv2.putText(o3,"Thresh_Tozero",(40,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),3,cv2.LINE_AA)
ret, o4 = cv2.threshold(gray_img_orig_clahe, th, max_val, cv2.THRESH_TOZERO_INV)
cv2.putText(o4,"Thresh_Tozero_inv",(40,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),3,cv2.LINE_AA)
ret, o5 = cv2.threshold(gray_img_orig_clahe, th, max_val, cv2.THRESH_TRUNC)
cv2.putText(o5,"Thresh_trunc",(40,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),3,cv2.LINE_AA)
ret ,o6=  cv2.threshold(gray_img_orig_clahe, th, max_val,  cv2.THRESH_OTSU)
cv2.putText(o6,"Thresh_OSTU",(40,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),3,cv2.LINE_AA)

final=np.concatenate((o1,o2,o3),axis=1)
final1=np.concatenate((o4,o5,o6),axis=1)

cv2.imwrite("Image1.jpg",final)
cv2.imwrite("Image2.jpg",final1)
```

```python

plt.subplot(211)
plt.imshow(final,cmap="gray")
plt.subplot(212)
plt.imshow(final1,cmap="gray")
plt.show()
```

**Step 9:** Adaptive Thresholding

```python
gray_image = cv2.imread('sample.JPG',0)
gray_image1 = cv2.imread('sample1.JPG',0)
thresh1 = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
thresh2 = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 31, 3)
thresh3 = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 13, 5)
thresh4 = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 4)

thresh11 = cv2.adaptiveThreshold(gray_image1, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
thresh12 = cv2.adaptiveThreshold(gray_image1, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 31, 5)
thresh13 = cv2.adaptiveThreshold(gray_image1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21,5 )
thresh14 = cv2.adaptiveThreshold(gray_image1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 5)

final=np.concatenate((thresh1,thresh2,thresh3,thresh4),axis=1)
final1=np.concatenate((thresh11,thresh12,thresh13,thresh14),axis=1)

cv2.imwrite('rect.jpg',final)
cv2.imwrite('rect1.jpg',final1)

```

```python
%matplotlib inline

plt.subplot(211)
plt.imshow(final,cmap="gray")
plt.subplot(212)
plt.imshow(final1,cmap="gray")
plt.show()
```

**Step 10:** OTSU Binarization

```python
gray_image = cv2.imread('sample.JPG',0)
gray_image1 = cv2.imread('sample1.JPG',0)
ret,thresh1 = cv2.threshold(gray_image,0, 255,  cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret,thresh2 = cv2.threshold(gray_image1,0, 255,  cv2.THRESH_BINARY+cv2.THRESH_OTSU)

cv2.imwrite('rect.jpeg',np.concatenate((thresh1,thresh2),axis=1))
```
